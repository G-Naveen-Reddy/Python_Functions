1. What is the difference between a function and a method in Python ?

 A method is a function that belongs to an object and is called using object.method().

 2. Explain the concept of function arguments and parameters in Python ?

 Arguments are the actual values passed to a function when calling it.

 4. What is the purpose of the `return` statement in a Python function ?
 - The return statement is used to send a value back from a function to the caller. It allows functions to produce an output that can be stored or used later. Without return, a function only performs an action but does not give back any result.

5. What are iterators in Python and how do they differ from iterables?
- Iterable: An object that can be looped over (e.g., lists, tuples, strings). It has an __iter__() method that returns an iterator.

 Iterator: An object that produces values one at a time using __next__(). It remembers its state and does not reset like an iterable.

 Difference:

 An iterable can be converted into an iterator, but an iterator can only be traversed once.










 6. Explain the concept of generators in Python and how they are defined ?
 - A generator is a special type of iterator that yields values one at a time using the yield keyword instead of return.

 How they are defined:

 Generators are created using functions with yield, or by using generator expressions. They produce values lazily, meaning they generate data only when needed, making them memory-efficient.

 7. What are the advantages of using generators over regular functions ?
 - Advantages of Generators over Regular Functions:
Memory Efficient – Generate values lazily, avoiding the need to store large data in memory.

 Faster Execution – No need to compute all values at once; produces values as needed.

 State Retention – Maintains state between yield calls, avoiding the need for manual tracking.

 Infinite Sequences – Can generate infinite data streams without running out of memory.

 Simpler Code – Eliminates the need for explicit iteration logic, making code more readable.

 9. Explain the purpose and usage of the `map()` function in Python ?
 - The map() function applies a given function to all items in an iterable and returns a map object (an iterator).

 Purpose:

 To transform elements of an iterable without using explicit loops.

 Usage:

 Used for applying a function to each element of a list, tuple, etc.

 Helps in performing element-wise operations efficiently.











 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python ?
 - map()
Purpose: Applies a given function to all items in an iterable (like a list) and returns a map object (iterator).

 Returns: An iterable of transformed elements.

 reduce()

 Purpose: Applies a given function to reduce the iterable to a single cumulative result (e.g., summing elements).

 Returns: A single value, the result of the reduction.

 filter()

 Purpose: Filters elements from an iterable based on a condition defined by a function.

 Returns: An iterable of elements that satisfy the condition.

 Summary:

 map() transforms all items.

 reduce() accumulates items into a single value.

 filter() selects items based on a condition.

```python
 '''
 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
'''
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
numbers=[1,2,3,4,5]
sum_even_numbers(numbers)
```

```python
 '''
 4. Write a Python function that checks if a given number is prime or not from 1 to 200
 '''
 def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
is_prime(5)
```

```python
 '''
 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms
'''
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.previous = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            if self.count == 0:
                self.count += 1
                return 0
            elif self.count == 1:
                self.count += 1
                return 1
            else:
                fib_number = self.previous + self.current
                self.previous = self.current
                self.current = fib_number
                self.count += 1
                return fib_number
        else:
            raise StopIteration
```

```python
'''
 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
'''
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
list(powers_of_2(5))
```

```python
'''
 7. Implement a generator function that reads a file line by line and yields each line as a string.
'''
def read_file_line_by_line(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
'''
 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
'''
tuples = [(1, 4), (2, 1), (3, 9), (4, 2)]

sorted_tuples = sorted(tuples, key=lambda x: x[1])

print(sorted_tuples)
```

```python
'''
 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
'''
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 25, 30, 35]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
'''
 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
'''
def remove_vowels(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, World!"
filtered_string = ''.join(filter(remove_vowels, input_string))

print(filtered_string)
```

```python
'''
11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €
'''
def calculate_order_value(orders):
    result = []
    for order in orders:
        order_number, price_per_item, quantity = order
        total_value = price_per_item * quantity
        if total_value < 100:
            total_value += 10
        result.append((order_number, total_value))
    return result


orders = [
    (1, 15, 5),
    (2, 25, 4),
    (3, 50, 1),
    (4, 30, 5)
]

order_values = calculate_order_value(orders)
print(order_values)
```
